```r
options(repr.plot.res = 300, repr.plot.width = 5, repr.plot.height = 2)
prison_df = read_csv("https://github.com/rfordatascience/tidytuesday/raw/master/data/2019/2019-01-22/prison_population.csv")
total_since_2011 = prison_df %>% filter(year >= 2011, year < 2016, pop_category %in% c("Male", "Female")) %>%
group_by(pop_category, year) %>%
summarize(n = sum(prison_population, na.rm = TRUE))
p = ggplot(total_since_2011, aes(x = year, y = n, color = pop_category))
p = p + geom_line()
p = p + labs(y= "Prision Population", x = "Year")
p = p + ggtitle("Prision Population by Sex from 2011-2015")
p = p + theme(plot.title = element_text(size = 8, face = "bold"))
p = p + theme(axis.title.y = element_text(size = 8))
p = p + theme(axis.title.x = element_text(size = 8))
p = p + theme(legend.title = element_text(size = 8))
p
```

The graph above shows the trend or sum of prision populations for both male and female from 2011-2015. Males are imprisioned far more than females. The number of males imprisionment started to decrease in 2014 while numbers of females have remained consistent over time. 

```r
options(repr.plot.res = 300, repr.plot.width = 5, repr.plot.height = 3)
prison_df = read_csv("https://github.com/rfordatascience/tidytuesday/raw/master/data/2019/2019-01-22/prison_population.csv")
total_df = prison_df %>% 
    filter(year == 2015, pop_category == "Total") %>%
    group_by(state) %>%
    summarize(prison_population = sum(prison_population, na.rm = TRUE)) %>%
    arrange(desc(prison_population)) %>%
    head(10)
p = ggplot(total_df, aes(x = prison_population, y = state %>% fct_inorder() %>% fct_rev()))
p = p + geom_col()
p = p + labs(y= "State", x = "Prision Population")
p = p + ggtitle("Top Ten States with the Highest Imprisionment in 2015")
p = p + theme(plot.title = element_text(size = 8, face = "bold"))
p = p + theme(axis.title.y = element_text(size = 8))
p = p + theme(axis.title.x = element_text(size = 8))
p
```

This chart ranks states with the highest prison population in 2015. 10 states are shown here with Texas, California and Florida making top 3. Texas prisionhad a population of about 150,000 people, California about 125,000 and Florida just about 100,000 people.

```r
options(repr.plot.res = 300, repr.plot.width = 5, repr.plot.height = 4)
prison_df = read_csv("https://github.com/rfordatascience/tidytuesday/raw/master/data/2019/2019-01-22/prison_population.csv")
total_df = prison_df %>% 
filter(year == 2015, !(pop_category %in% c("Total", "Female", "Male")), state %in% c("NY", "PA", "NC")) %>%
group_by(state, pop_category) %>%
summarize(prison_population = sum(prison_population, na.rm = TRUE)) %>%
arrange(desc(prison_population))
p = ggplot(total_df, aes(x = prison_population, fill = pop_category, y = state %>% fct_relevel("NY", "NC", "PA",) %>% fct_rev()))
p = p + geom_col()
p = p + labs(y = "State")
p = p + scale_fill_brewer(palette = "Dark2")
p = p + theme_dark()
p = p + labs(y= "State", x = "Prision Population")
p = p + ggtitle("New York, North Carolina & Pennsylvania Prision Populations by Race")
p = p + theme(plot.title = element_text(size = 8, face = "bold"))
p = p + theme(axis.title.y = element_text(size = 8))
p = p + theme(axis.title.x = element_text(size = 8))
p = p + theme(legend.title = element_text(size = 8))
p = p + theme(legend.position = "bottom")
p
```

This graph organizes the prision populations in Pennsylvania, New York, and North Carolina by race. As we can see, Black prisioners make up roughly about 50% of each state's population. Interestingly in Pennsylvania, there are low records for people who do not identify as Black or Latino. Those who identify as Other, Asian or Native American are amongst unlikey groups to be imprisioned due to their low population numbers.
