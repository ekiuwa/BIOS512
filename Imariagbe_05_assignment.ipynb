```r
volcano_df = read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-12/eruptions.csv")
df = volcano_df %>% group_by(volcano_name, end_year) %>%
count() %>%
arrange(desc(n)) 
```

Using `filter` and `arrange`, what is the year of the most recent eruption for the volcano above (i.e. the volcano with the most recorded eruptions in the data)? 

```r
df %>%
filter(!is.na(end_year))%>% 
arrange(desc(end_year)) 
```
